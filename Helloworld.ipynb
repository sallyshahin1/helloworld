```python
from flask import Flask
from pyngrok import ngrok

app = Flask(__name__)

@app.route('/')
def hello_world():
    return "Hello, World!"

if __name__ == '__main__':
    # Start ngrok tunnel
    public_url = ngrok.connect(5000)
    print("Public URL:", public_url)

    app.run(host='0.0.0.0', port=5000)
```
